```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import numpy as np

from SIR import *
from moonlight import *
from PID import *
```

# Free to evolve model

```python
sir = SIR()
t = np.arange(0, 365, 1)
gamma = - np.log(0.5/10)/365
u = 10*np.exp(-gamma * t)
S, I, R = sir.discrete_sim(u, t)
plt.plot(t, S, label = 'S')
plt.plot(t, I, label = 'I')
plt.plot(t, R, label = 'R')
#plt.yscale('log')
plt.legend()
plt.xlabel('days')
plt.savefig('free2evolve.png')
plt.show()
```

# PID Controller

```python

pid_controller = PID(sir, Kp=1, Ki=0, Kd=0)
S, I, R, u = pid_controller.pid(10**4, t)
plt.plot(t,I)
plt.plot(t, np.ones(t.shape)*10**4, linestyle='dashed')
plt.xlabel('days')
plt.ylabel('Infectious')
```
